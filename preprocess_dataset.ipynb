# Preprocess promoters

```python
df_promoters = pd.read_csv('datasets/df_regulonDB_locustag.csv')
```

```python
df_promoters.head(5)
```

```python
## select only promoters sigma70
df_promoters_70 = df_promoters[df_promoters['Sigma factor'] == 'sigma70']
df_promoters_70
```

```python
# Selection of genes controlled by a single promoter
df_promoters_70 = df_promoters_70[~df_promoters_70["TU first gene"].duplicated(keep=False)]
df_promoters_70
```

```python
# eliminate duplicated promoters
df_promoters_70 = df_promoters_70.drop_duplicates(subset=["Promoter sequence"], keep="first")
df_promoters_70
```

```python
# check if there is any duplicated promoter
df_promoters_70[df_promoters_70['Promoter ID'].duplicated()]
```

# Merge datasets

```python
df = pd.merge(df_promoters_70, df_tpm[['locus tag','median']], left_on="locus tag", right_on='locus tag')
```

```python
df
```

# Categorize Strength

```python
threshold_low = df['median'].max()/3
threshold_high = df['median'].max()*2/3

def categorize_median(median, low, high):
    if median <= low:
        return '0'
    elif median <= high:
        return '1'
    else:
        return '2'

df['Strength'] = df['median'].apply(categorize_median, args=(threshold_low, threshold_high))
```

```python
df
```

```python
df.to_csv('datasets/df_ML.csv', index=False)
```
