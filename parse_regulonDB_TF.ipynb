```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("K12/NetworkRegulatorGene.tsv", sep="\t", skiprows=24)
```

```python
df.head(2)
```

```python
dict_TF = {}

for i in range(len(df["5)regulatedName"])):
    if df["5)regulatedName"][i] not in dict_TF.keys():
        L = []
        L2 = []
        L2.append(df["1)regulatorId"][i])
        L2.append(df["2)regulatorName"][i])
        L2.append(df["6)function"][i])
        L2.append(df["7)confidenceLevel "][i])
        L.append(L2)
        dict_TF[df["5)regulatedName"][i]] = L
    else:
        L = []
        L.append(df["1)regulatorId"][i])
        L.append(df["2)regulatorName"][i])
        L.append(df["6)function"][i])
        L.append(df["7)confidenceLevel "][i])
        dict_TF[df["5)regulatedName"][i]].append(L)

```

```python
len(dict_TF.keys())
```

# Load promoters dataset

```python
df_promoters = pd.read_csv("dataframes/8_regulonDB_parsed_and_checked_locusTag.csv")
```

```python
df_promoters.head()
```

```python
def search_regulator(gene, dict):
    activator = False
    repressor = False
    if gene in dict.keys():
        for elem in dict[gene]:
            if elem[2] == "+":
                activator = True
            if elem[2] == "-":
                repressor = True
    return activator, repressor
```

```python
for i, row in df_promoters.iterrows():
    activator, repressor = search_regulator(row["TU first gene"], dict_TF)
    df_promoters.at[i, "activator"] = 1 if activator else 0
    df_promoters.at[i, "repressor"] = 1 if repressor else 0
```
