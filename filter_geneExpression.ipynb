```python
# Función para filtrar valores extremos usando IQR
def filter_outliers_iqr(row):
    # Calcular Q1, Q3 y el IQR
    Q1 = np.percentile(row, 25)
    Q3 = np.percentile(row, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Reemplazar valores extremos con NaN
    return row.where((row >= lower_bound) & (row <= upper_bound))

# Aplicar la función al DataFrame
filtered_df_tpm = tpm_wo_TF_curated.drop(columns="dataset id").apply(filter_outliers_iqr, axis=1)
```

# Convolutional Neural Network

```python
from keras.datasets import mnist
from keras.utils.np_utils import to_categorical
import tensorflow
import keras
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
def one_hot_encode_dna(sequences):
    """
    Codifica una lista de secuencias de ADN en una representación one-hot.

    Parameters:
    sequences (list of str): Lista de secuencias de ADN.

    Returns:
    np.array: Matriz 3D de secuencias codificadas one-hot.
    """
    # Definir el diccionario de codificación one-hot
    mapping = {
        'A': [1, 0, 0, 0],
        'C': [0, 1, 0, 0],
        'G': [0, 0, 1, 0],
        'T': [0, 0, 0, 1],
        'N': [0, 0, 0, 0]  # Para manejar nucleótidos desconocidos
    }
    
    # Obtener la longitud de la secuencia y el número de secuencias
    seq_length = len(sequences[0])
    num_sequences = len(sequences)
    
    # Crear una matriz vacía para almacenar las secuencias codificadas
    one_hot_encoded = np.zeros((num_sequences, seq_length, 4), dtype=np.int8)
    
    # Rellenar la matriz con los valores one-hot
    for i, seq in enumerate(sequences):
        for j, nucleotide in enumerate(seq):
            one_hot_encoded[i, j, :] = mapping.get(nucleotide, [0, 0, 0, 0])
    
    return one_hot_encoded
```

```python
df_for_cnn["sequence OHE CNN"] = df_for_cnn["Promoter sequence"].apply(lambda x: one_hot_encode_dna(x.upper()))
```

```python
sequences
```

```python
encoded_sequences = one_hot_encode_dna(sequences)
```

```python
labels = df_for_cnn["strength"]
```

```python
labels = np.array(labels).astype(np.float32).reshape(-1, 1)
```

```python
labels.shape
```

```python
labelencoder = LabelEncoder()
```

```python
labels = labelencoder.fit_transform(labels)
```

```python
labels
```

```python
labels = to_categorical(labels)
```

```python
labels
```

```python
labels.shape
```

```python
encoded_sequences.shape[1]
```

```python
encoded_sequences.shape[2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(encoded_sequences, labels, test_size=0.1, stratify=labels)
```

```python
encoded_sequences.shape[0]
```

```python
encoded_sequences.shape[1]
```

```python
encoded_sequences.shape[2]
```

```python
model = Sequential([
    Conv1D(filters=16, kernel_size=7, activation='relu', input_shape=(encoded_sequences.shape[1], encoded_sequences.shape[2])),
    Conv1D(filters=16, kernel_size=6, activation='relu', input_shape=(encoded_sequences.shape[1], encoded_sequences.shape[2])),
    Conv1D(filters=16, kernel_size=5, activation='relu', input_shape=(encoded_sequences.shape[1], encoded_sequences.shape[2])),
    MaxPooling1D(pool_size=4),
    Dropout(0.8),
    Flatten(),
    Dense(16, activation='relu'),
    Dense(5, activation='softmax')  # Salida única para clasificación binaria
])

# Compilar el modelo
model.compile(optimizer="rmsprop", loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Entrenar el modelo
model.fit(X_train, y_train, epochs=500, batch_size=32, validation_split=0.2)
```

```python
# Evaluar el modelo
loss_test, accuracy_test = model.evaluate(X_test, y_test)
print(f'Loss: {loss_test}, Accuracy: {accuracy_test}')
```

```python
def one_hot_encode_dna(sequences):
    """
    Codifica una lista de secuencias de ADN en una representación one-hot.

    Parameters:
    sequences (list of str): Lista de secuencias de ADN.

    Returns:
    np.array: Matriz 3D de secuencias codificadas one-hot.
    """
    # Definir el diccionario de codificación one-hot
    mapping = {
        'A': [1, 0, 0, 0],
        'C': [0, 1, 0, 0],
        'G': [0, 0, 1, 0],
        'T': [0, 0, 0, 1],
        'N': [0, 0, 0, 0]  # Para manejar nucleótidos desconocidos
    }
    
    # Obtener la longitud de la secuencia y el número de secuencias
    seq_length = len(sequences[0])
    num_sequences = len(sequences)
    
    # Crear una matriz vacía para almacenar las secuencias codificadas
    one_hot_encoded = np.zeros((num_sequences, seq_length, 4), dtype=np.int8)
    
    # Rellenar la matriz con los valores one-hot
    for i, seq in enumerate(sequences):
        for j, nucleotide in enumerate(seq):
            one_hot_encoded[i, j, :] = mapping.get(nucleotide, [0, 0, 0, 0])
    
    return one_hot_encoded
```
