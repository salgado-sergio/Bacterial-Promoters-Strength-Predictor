# Find the spacer region of promoters

```python
from Bio import SeqIO
from Bio.Seq import Seq

# Ruta al archivo FASTA
fasta_file = "K12\\NC_000913.fasta"

# Leer el archivo FASTA que contiene una única secuencia
for seq_record in SeqIO.parse(fasta_file, "fasta"):
    K12_genome = str(seq_record.seq)
```

```python
def extract_spacer(start35, end35, start10, end10, strand, genome):

    try:
        if strand == "forward":
            spacer = genome[int(end35):int(start10)-1]

        if strand == "reverse":
            spacer = genome[int(end10):int(start35)-1]
            seq = Seq(spacer)
            spacer = str(seq.reverse_complement())

        return spacer
    except:
        return ""
```

```python
df.head(1)
```

```python
df["spacer"] = df.apply(lambda row: extract_spacer(row["Promoter -35 box start"], row["Promoter -35 box end"], row["Promoter -10 box start"], row["Promoter -10 box end"], row["Operon strand"], K12_genome), axis=1)
```
