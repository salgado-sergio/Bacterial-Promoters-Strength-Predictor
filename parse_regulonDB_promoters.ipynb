```python
import pandas as pd
import json
import numpy as np
```

```python
### file regulondbdatamarts.operonDatamart.json downloaded from https://regulondb.ccg.unam.mx/manual/apiSoftware/docker

json_file = "databases/RegulonDB_v12/regulondbdatamarts.operonDatamart.json"

with open(json_file, 'r') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame({"Operon ID":[], 
                   "Operon name":[], 
                   "Operon strand": [], 
                   "TU ID": [], 
                   "TU name":[],
                   "TU first gene":[],                    
                   "Promoter ID":[], 
                   "Promoter name":[],                   
                   "Sigma factor":[], 
                   "Promoter sequence":[],
                   "Promoter -35 box start": [],
                   "Promoter -35 box end": [],
                   "Promoter -35 box sequence": [],
                   "Promoter -10 box start": [],
                   "Promoter -10 box end": [],
                   "Promoter -10 box sequence": [],                                 
                   "Promoter confidence": []}, dtype='object')

row = 0

for i in range(len(data)):
    for elem2 in data[i]['transcriptionUnits']:
        df.loc[row, "Operon ID"] = data[i]['_id']
        df.loc[row, "Operon name"] = data[i]['operon']['name']
        df.loc[row, "Operon strand"] = data[i]['operon']['strand']
        df.loc[row, "TU ID"] = elem2['_id']
        
        try:
            try:
                df.loc[row, "TU name"] = elem2['name']
            except KeyError:
                pass

            try:
                df.loc[row, "TU first gene"] = elem2['firstGene']['name']
            except KeyError:
                pass

            try:
                df.loc[row, "Promoter ID"] = elem2['promoter']['_id']
            except KeyError:
                pass

            try:
                df.loc[row, "Promoter name"] = elem2['promoter']['name']
            except KeyError:
                pass

            try:
                df.loc[row, "Sigma factor"] = elem2['promoter']['bindsSigmaFactor']['abbreviatedName']
            except KeyError:
                pass

            try:
                df.loc[row, "Promoter sequence"] = elem2['promoter']['sequence']
            except KeyError:
                pass

            try:
                df.loc[row, "Promoter -35 box start"] = elem2['promoter']['boxes'][1]["leftEndPosition"]
            except (KeyError, IndexError):
                pass

            try:
                df.loc[row, "Promoter -35 box end"] = elem2['promoter']['boxes'][1]["rightEndPosition"]
            except (KeyError, IndexError):
                pass

            try:
                df.loc[row, "Promoter -35 box sequence"] = elem2['promoter']['boxes'][1]["sequence"]
            except (KeyError, IndexError):
                pass

            try:
                df.loc[row, "Promoter -10 box start"] = elem2['promoter']['boxes'][0]["leftEndPosition"]
            except (KeyError, IndexError):
                pass

            try:
                df.loc[row, "Promoter -10 box end"] = elem2['promoter']['boxes'][0]["rightEndPosition"]
            except (KeyError, IndexError):
                pass

            try:
                df.loc[row, "Promoter -10 box sequence"] = elem2['promoter']['boxes'][0]["sequence"]
            except (KeyError, IndexError):
                pass

            try:
                df.loc[row, "Promoter confidence"] = elem2['promoter']['confidenceLevel']
            except KeyError:
                pass

        except Exception as e:
            print(f"Error procesando la fila {row}: {e}")

        row += 1
```
