# Loading data

```python
df = pd.read_csv("iModulon\\precise1k-master\\data\\precise1k\\log_tpm_qc.csv")
```

```python
df_transposed = df.transpose()
```

```python
df_transposed.columns = df["Unnamed: 0"]
```

```python
df_transposed.reset_index(inplace=True)
```

```python
df_transposed_int.iloc[:,1:] = df_transposed.iloc[:,1:].astype('float32')
```

# Filtering experiments

## Load metadata

```python
df_metadata = pd.read_csv("iModulon\precise1k-master\data\precise1k\\metadata_qc.csv")
```

## Filter dataset

```python
dataset_list = df_metadata[df_metadata["Strain Description"]=="Escherichia coli K-12 MG1655"]["Unnamed: 0"]
```

```python
dataset_list = list(dataset_list)
```

```python
dataset_list
```

```python
df_transposed_int_filtered = df_transposed_int[df_transposed_int["index"].isin(dataset_list)].reset_index(drop=True)
```

```python
df_transposed_int_filtered.columns.drop('index')
```

```python
df_transposed_int_filtered[df_transposed_int_filtered.columns.drop('index')] = df_transposed_int_filtered[df_transposed_int_filtered.columns.drop('index')].astype('float32')
```
