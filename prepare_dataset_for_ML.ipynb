# Gene Expression

```python
df_geneExpression = pd.read_csv("dataframes/9_geneExpressionK12.csv")
```

```python
df_geneExpression.drop(columns="b4711", inplace=True)
```

```python
df_geneExpression_statistics = pd.DataFrame({"locus tag" : df_geneExpression.drop(columns="index").columns,
                                            "std_dev" : df_geneExpression.drop(columns="index").std().values,
                                            "cv" : df_geneExpression.drop(columns="index").std().values / df_geneExpression.drop(columns="index").mean().values})
```

```python
df_geneExpression_statistics
```

```python
from sklearn.cluster import KMeans

# Seleccionar la métrica de variabilidad (por ejemplo, desviación estándar o coeficiente de variación)
X = df_geneExpression_statistics[['cv']]

# Aplicar K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

# Añadir las etiquetas de los clusters al DataFrame original
df_geneExpression_statistics['cluster cv'] = kmeans.labels_
```

```python
threshold_low = df_geneExpression_statistics['cv'].quantile(0.33)
threshold_high = df_geneExpression_statistics['cv'].quantile(0.66)

print(f"Umbral bajo (percentil 33): {threshold_low}")
print(f"Umbral alto (percentil 66): {threshold_high}")

# Función para categorizar en función de CV
def categorize_cv(cv, low, high):
    if cv <= low:
        return 'bajo'
    elif cv <= high:
        return 'medio'
    else:
        return 'alto'

# Aplicar la función de categorización
df_geneExpression_statistics['cv_category'] = df_geneExpression_statistics['cv'].apply(categorize_cv, args=(threshold_low, threshold_high))
```

```python
df_geneExpression_statistics
```

```python
print(df_geneExpression_statistics[df_geneExpression_statistics["cluster cv"]==0].min()["cv"])
print(df_geneExpression_statistics[df_geneExpression_statistics["cluster cv"]==0].max()["cv"])
```

```python
print(df_geneExpression_statistics[df_geneExpression_statistics["cluster cv"]==1].min()["cv"])
print(df_geneExpression_statistics[df_geneExpression_statistics["cluster cv"]==1].max()["cv"])
```

```python
print(df_geneExpression_statistics[df_geneExpression_statistics["cluster cv"]==2].min()["cv"])
print(df_geneExpression_statistics[df_geneExpression_statistics["cluster cv"]==2].max()["cv"])
```

```python
# Rename of K-means clusters

df_geneExpression_statistics["cluster cv"] = df_geneExpression_statistics["cluster cv"].apply(lambda x: "low" if x==1 else x)
df_geneExpression_statistics["cluster cv"] = df_geneExpression_statistics["cluster cv"].apply(lambda x: "medium" if x==0 else x)
df_geneExpression_statistics["cluster cv"] = df_geneExpression_statistics["cluster cv"].apply(lambda x: "high" if x==2 else x)

df_geneExpression_statistics["cluster cv"] = df_geneExpression_statistics["cluster cv"].apply(lambda x: 0 if x=="low" else x)
df_geneExpression_statistics["cluster cv"] = df_geneExpression_statistics["cluster cv"].apply(lambda x: 1 if x=="medium" else x)
df_geneExpression_statistics["cluster cv"] = df_geneExpression_statistics["cluster cv"].apply(lambda x: 2 if x=="high" else x)
```
